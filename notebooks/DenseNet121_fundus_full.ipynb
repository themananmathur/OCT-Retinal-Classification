```python
import tensorflow
import os,cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from keras import backend as K
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Convolution2D, MaxPooling2D
from keras.optimizers import SGD,RMSprop,adam
from keras import regularizers

import os
os.chdir('D:\\Dataset_fundus\\Jenysir')

#train_x = np.load('trainX.npy')
#train_y = np.load('trainY.npy')
train_x  = np.load('x_train1.npy')
test_x = np.load('x_test2.npy')
train_y = np.load('y_train4.npy')
test_y = np.load('y_test2.npy')

X_train=train_x
X_test=test_x

y_train=train_y
y_test=test_y

from sklearn.model_selection import train_test_split

# Split the data
x_train_x, x_valid, y_train_x, y_valid = train_test_split(X_train,y_train, test_size=0.2, shuffle= True)


#X_train = X_train.reshape(X_train.shape[0],512,496,1).astype('float32')/255

#X_test  = X_test.reshape(X_test.shape[0],512,496,1).astype('float32')/255

X_train_x = x_train_x.reshape(x_train_x.shape[0],256,256,3).astype('float32')

X_test  = X_test.reshape(X_test.shape[0],256,256,3).astype('float32')

X_valid = x_valid.reshape(x_valid.shape[0],256,256,3).astype('float32')



#X_val = X_val.reshape(X_val.shape[0],300,300,3).astype('float32')/255


#import tensorflow as tf
#keras
import keras
from keras_applications.densenet import DenseNet121

model = DenseNet121(include_top = False,
                    input_shape=(None,None,3),
                  weights=None,
                  backend=keras.backend,
                  layers=keras.layers,
                  models=keras.models,
                  utils=keras.utils)
model.summary()

from keras.layers import *

x1=model.output
x2=GlobalAveragePooling2D()(x1)
x3=Dropout(0.3)(x2)
out1=Dense(10,activation='relu')(x3)
out2=Dense(2,activation='softmax')(out1)

from keras import Model
model=Model(inputs=model.input,outputs=out2) ### define keras' functional model
model.summary()




import os
os.chdir('D:\\Dataset_fundus\\Jenysir\\Models')
from keras.callbacks import ModelCheckpoint,TensorBoard,CSVLogger,ReduceLROnPlateau,LearningRateScheduler,EarlyStopping
mc = ModelCheckpoint('./abn_Dense_Train4_250.h5', monitor='val_loss', mode='min',save_best_only=True)
es = EarlyStopping(monitor='val_loss', min_delta=0, patience=25, verbose=0, mode='auto', baseline=None, restore_best_weights=True)
cv = CSVLogger('./Dense_Train4_250.csv',append=True)

import keras
adam = keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, decay=0.0, amsgrad=False)


model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])

# we can compare the performance with or without data augmentation

batch_size = 4
epochs = 250

history = model.fit(X_train_x,y_train_x,batch_size=batch_size,epochs=epochs,validation_data=(x_valid, y_valid),verbose = 1,callbacks=[es,mc,cv],shuffle=True)
```
